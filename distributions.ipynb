```python
import matplotlib.pyplot as plt #Import the plot library
import numpy as np
%matplotlib inline 
%config InlineBackend.figure_format = 'svg'

mu=0.5
sigma=0.5

uniform_low=0.25
uniform_high=0.8


uniform = np.random.uniform(uniform_low, uniform_high, 10000)
plt.hist(uniform, 50, density=1)

plt.savefig("uniform.svg")

plt.figure()

distro = np.random.normal(mu, sigma, 10000)/0.02
plt.hist(distro, 50, density=1)

plt.figure()

distro2 = np.random.logistic(mu, sigma, 10000)/0.02
plt.hist(distro2, 50, density=1)

distro = np.random.normal(mu, sigma, 10000)/0.02
plt.hist(distro, 50, density=True)

plt.figure()
logistic_cumulative = np.random.logistic(mu, sigma, 10000)/0.02
plt.hist(logistic_cumulative, 50, density=1, cumulative=True)
plt.savefig("logistic_cumulative.svg")

plt.figure()
distro = np.random.binomial(1, .6, 10000)/0.5
plt.hist(distro, 2 , density=1)
plt.savefig("bernoulli.svg")

plt.figure()
distro = np.random.binomial(100, .6, 10000)/0.01
plt.hist(distro, 100 , density=1)
plt.savefig("binomial.svg")
```
