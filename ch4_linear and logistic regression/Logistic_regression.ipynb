## Practical example: Cardiac disease modeling with logistic regression

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn import linear_model
import seaborn as sns
%matplotlib inline
import matplotlib.pyplot as plt
sns.set(style='whitegrid', context='notebook')
```

```python
df = pd.read_csv("C:/Users/이태호/Desktop/대학교/3학년/인공지능/개발자를 위한 머신 러닝/Machine-Learning-for-Developers-code-updates/Chapter04/data/CHD.csv", header=0, engine='python')
plt.figure() # Create a new figure
plt.axis ([0,70,-0.2,1.2])
plt.title('Original data')
plt.scatter(df['age'],df['chd']) #Plot a scatter draw of the random datapoints
```

```python
logistic = linear_model.LogisticRegression(C=1e5)
logistic.fit(df['age'].values.reshape(-1, 1), df['chd'].values.reshape(-1, 1).ravel())
```

```python
x_plot = np.linspace(10, 90, 100)
oneprob = []
zeroprob =[]
predict = []
plt.figure(figsize=(10,10))

for i in x_plot:
   oneprob.append (logistic.predict_proba(np.array([[i]]))[0][1]);
   zeroprob.append (logistic.predict_proba(np.array([[i]]))[0][0]);
   predict.append (logistic.predict(np.array([[i]]))[0]);

plt.plot(x_plot, oneprob);
plt.plot(x_plot, zeroprob)
plt.plot(x_plot, predict);
plt.scatter(df['age'], df['chd'])
```
