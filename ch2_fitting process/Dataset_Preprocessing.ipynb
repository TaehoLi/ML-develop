# normalization and feature scaling

```python
from sklearn import preprocessing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv("C:/Users/이태호/Desktop/대학교/3학년/2학기/인공지능/개발자를 위한 머신 러닝/Machine-Learning-for-Developers-code-updates/Chapter02/Code/data/mpg.csv"
                , engine='python')
plt.figure(figsize=(10,8))
print(df.columns)
partialcolumns = df[['acceleration', 'mpg']]
std_scale = preprocessing.StandardScaler().fit(partialcolumns)
df_std = std_scale.transform(partialcolumns)
plt.scatter(partialcolumns['acceleration'], partialcolumns['mpg'], color='gray', marker='^')
plt.scatter(df_std[:,0], df_std[:,1])
```
