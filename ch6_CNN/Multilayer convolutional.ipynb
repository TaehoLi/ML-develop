# Deep Learning with Keras

```python
from keras.models import Model
from keras.preprocessing import image
from keras.optimizers import SGD
from keras.applications.inception_v3 import InceptionV3, decode_predictions, preprocess_input

import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
model=InceptionV3()
model.compile(optimizer=SGD(), loss='categorical_crossentropy')
```

```python
# resize into VGG16 trained images' format
import os
os.chdir('C:/Users//Desktop//3//   /Machine-Learning-for-Developers-code-updates/Chapter06')

im = cv2.resize(cv2.imread('blue_jay.jpg').astype(float), (299, 299))
im = np.expand_dims(im, axis=0)
im = im / 255.
im = im - 0.5
im =  im * 2
plt.figure (figsize=(10,10))
plt.imshow(im[0], cmap=plt.get_cmap('binary_r'))
plt.show()
```

```python
out = model.predict(im)
print('Predicted:', decode_predictions(out, top=3)[0])
print (np.argmax(out))
```
